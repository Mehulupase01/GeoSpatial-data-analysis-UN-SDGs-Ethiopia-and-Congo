## Task 3: Spatial Lag & Moran's Plot

In Lab 4 (http://darribas.org/gds16/labs/Lab_04.html) by Dani Arribas, you can see examples of weight matrices and at the end an explanation of spatial lag and the Morans plot (Watch out, this is not Moran's I/ Moran's spatial autocorrelation, which is also often calculated in spatial analytics), We have provided a modified version of this lab ready to go in this repo. 
Read and understand the concept of spatial lag first. In a Moran's plot, you can show how the value of a variable relates to it's spatial lag (or the average of this value in the surrounding areas). 

Additional information: 
* [Dani Arribas written course material](http://darribas.org/gds_scipy16/)
* [installing PySAL](https://pysal.org/docs/install/) and [PySAL tutorial](https://pysal.org/libpysal/tutorial.html)

Your tasks:

Create two Morans plots based on the variables related to your research questions and interpret them: 

What can you say about the entire continent?

Can you find meaningful subregions? 

What if you use block weights around those subregions and re-run the analysis? 

```python
import pandas as pd
import geopandas as gpd
import libpysal as ps
import esda
import matplotlib.pyplot as plt
import seaborn as sns

congo_shapefile_path = "/workspaces/assignment-2-Mehulupase01/gadm41_COD_shp/gadm41_COD_1.shp"
ethiopia_shapefile_path = "/workspaces/assignment-2-Mehulupase01/gadm41_ETH_shp/gadm41_ETH_1.shp"
congo_r1_path = "/workspaces/assignment-2-Mehulupase01/Congo-R1.xls"
congo_r2_path = "/workspaces/assignment-2-Mehulupase01/Congo-R2.xls"
ethiopia_r1_path = "/workspaces/assignment-2-Mehulupase01/Ethiopia-R1.xls"
ethiopia_r2_path = "/workspaces/assignment-2-Mehulupase01/Ethiopia-R2.xls"

congo_gdf = gpd.read_file(congo_shapefile_path).explode(index_parts=True).reset_index(drop=True)
ethiopia_gdf = gpd.read_file(ethiopia_shapefile_path).explode(index_parts=True).reset_index(drop=True)

congo_r1_data = pd.read_excel(congo_r1_path).melt(id_vars=["Country Name"], var_name="Year", value_name="Employment_Poverty")
congo_r2_data = pd.read_excel(congo_r2_path).melt(id_vars=["Country Name"], var_name="Year", value_name="Undernourishment")
ethiopia_r1_data = pd.read_excel(ethiopia_r1_path).melt(id_vars=["Country Name"], var_name="Year", value_name="Employment_Poverty")
ethiopia_r2_data = pd.read_excel(ethiopia_r2_path).melt(id_vars=["Country Name"], var_name="Year", value_name="Undernourishment")

congo_r1_data['Year'] = congo_r1_data['Year'].astype(int)
ethiopia_r1_data['Year'] = ethiopia_r1_data['Year'].astype(int)
congo_r1_data = congo_r1_data[(congo_r1_data['Year'] >= 1960) & (congo_r1_data['Year'] <= 2023)]
ethiopia_r1_data = ethiopia_r1_data[(ethiopia_r1_data['Year'] >= 1960) & (ethiopia_r1_data['Year'] <= 2023)]

congo_r2_data['Year'] = congo_r2_data['Year'].astype(int)
ethiopia_r2_data['Year'] = ethiopia_r2_data['Year'].astype(int)
congo_r2_data = congo_r2_data[(congo_r2_data['Year'] >= 2001) & (congo_r2_data['Year'] <= 2021)]
ethiopia_r2_data = ethiopia_r2_data[(ethiopia_r2_data['Year'] >= 2001) & (ethiopia_r2_data['Year'] <= 2021)]

congo_r1_data['Country'] = 'Congo'
congo_r2_data['Country'] = 'Congo'
ethiopia_r1_data['Country'] = 'Ethiopia'
ethiopia_r2_data['Country'] = 'Ethiopia'

employment_data = pd.concat([congo_r1_data, ethiopia_r1_data], ignore_index=True)
undernourishment_data = pd.concat([congo_r2_data, ethiopia_r2_data], ignore_index=True)

congo_gdf['Country'] = 'Congo'
ethiopia_gdf['Country'] = 'Ethiopia'
africa_gdf = pd.concat([congo_gdf, ethiopia_gdf], ignore_index=True)
africa_gdf = africa_gdf.merge(employment_data, on='Country')
africa_gdf = africa_gdf.merge(undernourishment_data, on=['Country', 'Year'])

w = ps.weights.Queen.from_dataframe(africa_gdf)
w.transform = 'r'

africa_gdf['w_employment'] = ps.weights.spatial_lag.lag_spatial(w, africa_gdf['Employment_Poverty'])
africa_gdf['w_undernourishment'] = ps.weights.spatial_lag.lag_spatial(w, africa_gdf['Undernourishment'])

fig, ax = plt.subplots(1, 1, figsize=(8, 6))
sns.regplot(x='Employment_Poverty', y='w_employment', data=africa_gdf, ci=None, ax=ax)
ax.set_title("Moran's Plot: Employment Rate vs. Spatial Lag")
ax.set_xlabel("Employment Rate")
ax.set_ylabel("Spatial Lag of Employment Rate")
plt.show()

fig, ax = plt.subplots(1, 1, figsize=(8, 6))
sns.regplot(x='Undernourishment', y='w_undernourishment', data=africa_gdf, ci=None, ax=ax)
ax.set_title("Moran's Plot: Undernourishment vs. Spatial Lag")
ax.set_xlabel("Undernourishment")
ax.set_ylabel("Spatial Lag of Undernourishment")
plt.show()

subregion_gdf = africa_gdf[africa_gdf['NAME_1'].isin(['Kinshasa', 'Oromia'])]

block_w = ps.weights.DistanceBand.from_dataframe(subregion_gdf, threshold=500000, binary=True)
block_w.transform = 'r'

subregion_gdf['w_employment_block'] = ps.weights.spatial_lag.lag_spatial(block_w, subregion_gdf['Employment_Poverty'])
subregion_gdf['w_undernourishment_block'] = ps.weights.spatial_lag.lag_spatial(block_w, subregion_gdf['Undernourishment'])


fig, ax = plt.subplots(1, 1, figsize=(8, 6))
sns.regplot(x='Employment_Poverty', y='w_employment_block', data=subregion_gdf, ci=None, ax=ax)
ax.set_title("Moran's Plot: Employment Rate (Subregions) vs. Block Spatial Lag")
ax.set_xlabel("Employment Rate")
ax.set_ylabel("Block Spatial Lag of Employment Rate")
plt.show()

fig, ax = plt.subplots(1, 1, figsize=(8, 6))
sns.regplot(x='Undernourishment', y='w_undernourishment_block', data=subregion_gdf, ci=None, ax=ax)
ax.set_title("Moran's Plot: Undernourishment (Subregions) vs. Block Spatial Lag")
ax.set_xlabel("Undernourishment")
ax.set_ylabel("Block Spatial Lag of Undernourishment")
plt.show()
```

***Interpretation:***

What can you say about the whole continent?

- The Moran scatter plot of the employment rates shows strong positive spatial autocorrelation; there are two clear-cut clusters: one falls at a low level of employment, around 55-60%, whereas the second is at a high level, approximately 80-85%. This would confirm that regions with comparable employment conditions—that is, equally low and high respectively—are geographically proximate to one another. Undernourishment is more dispersed in the plot with relatively weaker positive spatial autocorrelation. Without which, it would have inferred that the levels would be similar between near regions. Values range from lower undernourishment levels at around 15-20% and go up to higher undernourishment levels at 35-45%.

Can you find meaningful subregions?

- The analysis shows the possible sub-regions characterized by homogeneity in socio-economic conditions. For example: That cluster has lower rates of employment, probably something about 55 to 60 percent, indicating rural or economically backward areas, maybe some provinces in the DRC. This higher cluster of employment could be at about 80-85%, which may indicate the presence of an urbanized or economically developed region; such would be areas in Ethiopia that have better cases of job availability.

- For undernourishment, the high values in some regions could show places that have food insecurity, likely due to specific drought reasons resulting from small agricultural yield; an example could be part of the country of Ethiopia.

- However, the unit of analysis being at the country level, the more accurate sub-regions are inferred based on country-wide data rather than detailed local data.

What if you make use of block weights around those subregions and run the analysis once more?

- The application of the block weight for subregions yields more local features, such as Kinshasa in DRC and Oromia in Ethiopia. A Moran's plot for subregions for employment rates using block weights shows a more complex relationship that might be negative in some places for a spatial effect. That may be so that the local conditions of one area influence the neighbors differently. For undernourishment, the block-weighted local plot has localized clusters but somewhat less uniformity than full continent analysis. This may suggest that areas of higher or lower levels of undernourishment may have distinct localized interaction. This helps to begin giving a more localized focus on the "hotspots," or localized areas of concern, and conveys how diversified conditions at the local level can be from larger national trends. Hence, policies may need to be tailored for certain regions to work out. All in all, new plots show that block weights can also provide a finer distinction about regional spatial patterns that capture important deviations within countries not perceived for the whole continent.
